```python
df = pd.read_csv("MarketTop2000.csv")
```

# Most capital is utilised in agrregate pharmaceautical sector of around 100 crores 

```python
## Top 10 large cap companies with capital

dfsegment = df.groupby('Name')['Market Cap'].sum().reset_index().sort_values(by='Market Cap', ascending=False)
```

```python
df_large = dfsegment[dfsegment['Market Cap'] > 20000]
```

```python
datalarge = df_large.set_index('Name', inplace=True)
dfbig = df_large.head(10)
```

```python
#dfbig.plot(kind="pie", autopct='%1.1f%%', y = 'Market Cap', shadow=True,  legend=True, title='Frequency of sector ', ylabel='', labeldistance=None,
#                 colors = ['red', 'pink', 'steelblue','black','yellow','grey','brown','green','orange','purple'])
#dfbig.plot(kind='bar', title='Frequency of Sector',ylabel='capital in crores', xlabel='sector', figsize=(25, 7))
plt.rc('xtick', labelsize=10) 
plt.rc('ytick', labelsize=10)
plt.rc('font', size=10) 
colors = ['red', 'pink', 'steelblue','black','yellow','grey','brown','green','orange','purple']
data = dfbig["Market Cap"]
ax = data.plot(kind="pie",y = 'Market Cap' ,autopct='%1.1f%%', shadow=True, colors=colors, legend=True, title='top large cap', ylabel='', labeldistance=None)
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()
```
